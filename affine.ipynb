```python
%matplotlib inline
import numpy as np
from parse import parse
from matplotlib import pyplot as plt
```

```python
data, labels = parse('mnist_train.csv')
```

```python
# sanity check
fig, axs = plt.subplots(4,10)
for i in range(10):
    samples_of_digit_i = data[labels==i]
    count = samples_of_digit_i.shape[0]
    random_sample = samples_of_digit_i[np.random.choice(count, 4)]
    for j in range(4):
        axs[j,i].axis('off')
        axs[j,i].imshow(random_sample[j])
plt.show()
```
